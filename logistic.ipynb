```python
import pandas as pd
import numpy as np
import sklearn.linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance_data.csv')
```

```python
df
```

```python
plt.scatter(df.age,df.bought_insurance,marker='H')
plt.plot()
```

```python
model = sklearn.linear_model.LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x,x_test,y,y_test = train_test_split(df['age'],df['bought_insurance'],test_size=0.1)
```

```python
x_test
```

```python
model.fit(np.array(x).reshape(-1,1),y)
```

```python
model.predict([[val] for val in x_test])
```

```python
model.score(np.array(x).reshape(-1,1),y)
```

```python
model.predict([[val] for val in x_test])
```

```python
model.coef_
model.intercept_
```

```python
df['fixed'] = df['age']*model.coef_[0] + model.intercept_
df
```

```python
plt.scatter(df.age,df.bought_insurance,marker='H')
plt.plot(df['age'],df['fixed'])
```
